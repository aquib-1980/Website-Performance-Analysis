```python
plt.figure(figsize=(10,5))
df.groupby('DateHour')[['Sessions','Users']].sum().plot(ax=plt.gca())
plt.title('Sessions and users over time')
plt.xlabel('DateHour')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.bar(df['channel group'],df['Users'],color='orange')
plt.title('Total users by channel')
plt.xlabel('channel')
plt.ylabel('Users')
plt.show()#Direct channel has engaged more users
```

```python
df.head()
```

```python
plt.figure(figsize=(15,8))
plt.bar(df['channel group'], df['Average engagement time per session'])
plt.title('Average engagement time by channel')
plt.xlabel('Channel')
plt.ylabel('Avg time')
plt.show()#Organic vdos are the most that people watch
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='channel group', y='Engagement rate', data=df, palette='coolwarm')
plt.title('Engagement Rate distribution by channel')
plt.xlabel('channel')
plt.ylabel('Engagement Rate')
plt.show()#Referral channel has the most Engagement Rate
```

```python
grouped = df.groupby('channel group')[['Sessions', 'Engaged sessions']].sum()
grouped['Non-engaged sessions'] = grouped['Sessions'] - grouped['Engaged sessions']
print(grouped)
```

```python
grouped[['Engaged sessions', 'Non-engaged sessions']].plot(
    kind='bar', figsize=(10,6), color=['skyblue', 'lightcoral']
)
plt.title('Engaged vs Non-Engaged Sessions by Channel')
plt.xlabel('Channel Group')
plt.ylabel('Number of Sessions')
plt.xticks(rotation=45)
plt.show()#organic social has the most engaged sessions
```

```python
df_plot = df.groupby('DateHour')[['Engagement rate','Sessions']].mean().reset_index()
plt.figure(figsize=(20,10))
plt.plot(df_plot['DateHour'], df_plot['Engagement rate'], label='Engagemnt Rate', color='green')
plt.plot(df_plot['DateHour'], df_plot['Sessions'], label='Sessions', color='purple')
plt.title('Engagement Rate vs Sessions Over Time')
plt.xlabel('DateHour')
plt.legend()
plt.show()#Engagement Rate is high as compared to Sessions
```
